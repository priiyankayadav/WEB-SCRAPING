# 5. Scrape the details of trending repositories on Github.com.

Url = https://github.com/
    
You have to find the following details: A) Repository title B) Repository description C) Contributors count D) Language used

```python
from selenium.common.exceptions import ElementNotInteractableException
```

```python
#Connecting to web driver
driver = webdriver.Chrome(r"C:\Users\Administrator\Downloads\chromedriver_win32\chromedriver.exe")
driver.maximize_window()
```

```python
#Opening GitHub page
driver.get('https://github.com/trending')
driver.implicitly_wait(5)
```

```python
#Creating empty lists for storing scraped data
Rep_Title=[]
Rep_Desc=[] 
CC=[] 
Lang=[]
urls=[]
```

```python
#Collecting the information of top trending repos. from GitHub
rep_title=driver.find_elements_by_xpath('//article[@class="Box-row"]/h1/a')
desc=driver.find_elements_by_xpath('//article[@class="Box-row"]/p')
for i in [rep_title,desc]:
    for j in i:
        if i ==rep_title:
            Rep_Title.append(j.text)
            urls.append(j.get_attribute('href'))
        if i==desc:
            Rep_Desc.append(j.text)
for url in urls:
    driver.get(url)
    driver.implicitly_wait(3)
    div_list=driver.find_elements_by_xpath('//div[@class="BorderGrid BorderGrid--spacious"]/div')
    try:
        CC.append(((div_list[-2].text).split())[1])
    except:
        CC.append('-----')
    try:
        Lang.append(((div_list[-1].text).split())[1::2])
    except:
        Lang.append('-----')
```

```python
#Creating Dataframe for storing the collected data present in lists
Git_Tre = pd.DataFrame({})
Git_Tre['Repository Title']=Rep_Title[0:24]
Git_Tre['Repository Description']=Rep_Desc[0:24]
Git_Tre['Contributors']=CC[0:24]
Git_Tre['Language Used']=Lang[0:24]
```
